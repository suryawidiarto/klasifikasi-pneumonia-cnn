```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n_data64 = pd.read_csv('normal/64x64/training_history_pred_conf.csv', sep=',')
n_data128 = pd.read_csv('normal/128x128/training_history_pred_conf.csv', sep=',')
n_data256 = pd.read_csv('normal/256x256/training_history_pred_conf.csv', sep=',')

c_data64 = pd.read_csv('clahe/64x64/training_history_pred_conf.csv', sep=',')
c_data128 = pd.read_csv('clahe/128x128/training_history_pred_conf.csv', sep=',')
c_data256 = pd.read_csv('clahe/256x256/training_history_pred_conf.csv', sep=',')

plt.figure(figsize=(8,6),dpi=100)

NDat64 = n_data64['accuracy']
NDat128 = n_data128['accuracy']
NDat256 = n_data256['accuracy']

CDat64 = c_data64['accuracy']
CDat128 = c_data128['accuracy']
CDat256 = c_data256['accuracy']

labels = n_data64['Epoch']

plt.plot(NDat64, linestyle='dashed', marker='o')
plt.plot(NDat128, linestyle='dashed', marker='o')
plt.plot(NDat256, linestyle='dashed', marker='o')

plt.plot(CDat64, linestyle='solid', marker='o')
plt.plot(CDat128, linestyle='solid', marker='o')
plt.plot(CDat256, linestyle='solid', marker='o')

plt.title('Training Accuracy')
plt.ylabel('Accuracy')
plt.xticks(n_data64.index,labels, rotation=45)
plt.legend(['normal 64x64','normal 128x128','normal 256x256','clahe 64x64','clahe 128x128','clahe 256x256'], loc='best')
    
plt.savefig('training_accuracy.jpg')
plt.show()
```

```python
import numpy as np
data_normal = pd.read_csv('rata_rata_normal_train_val_test.csv', sep=',')
data_clahe = pd.read_csv('rata_rata_clahe_train_val_test.csv', sep=',')

plt.figure(figsize=(10,8),dpi=100)

TA_NData = data_normal['Train Acc']
TA_CData = data_clahe['Train Acc']
VA_NData = data_normal['Val Acc']
VA_CData = data_clahe['Val Acc']
TL_NData = data_normal['Train Loss']
TL_CData = data_clahe['Train Loss']
VL_NData = data_normal['Val Loss']
VL_CData = data_clahe['Val Loss']
TAC_NData = data_normal['Test Acc']
TAC_CData = data_clahe['Test Acc']

plot1 = plt.subplot2grid((3, 3), (0, 0))
plot2 = plt.subplot2grid((3, 3), (0, 1))
plot3 = plt.subplot2grid((3, 3), (1, 0))
plot4 = plt.subplot2grid((3, 3), (1, 1))
plot5 = plt.subplot2grid((3, 3), (0, 2))

# Training Accuracy
plot1.plot(TA_NData, linestyle='dashed', marker='o')
plot1.plot(TA_CData, linestyle='solid', marker='o')
plot1.set_title('Training Accuracy')
plot1.legend(['Normal','CLAHE'], loc='lower center')

# Validation Accuracy
plot2.plot(VA_NData, linestyle='dashed', marker='o')
plot2.plot(VA_CData, linestyle='solid', marker='o')
plot2.set_title('Validation Accuracy')
plot2.legend(['Normal','CLAHE'], loc='lower center')

# Training Loss
plot3.plot(TL_NData, linestyle='dashed', marker='o')
plot3.plot(TL_CData, linestyle='solid', marker='o')
plot3.set_title('Training Loss')
plot3.legend(['Normal','CLAHE'], loc='upper center')

# Validation Loss
plot4.plot(VL_NData, linestyle='dashed', marker='o')
plot4.plot(VL_CData, linestyle='solid', marker='o')
plot4.set_title('Validation Loss')
plot4.legend(['Normal','CLAHE'], loc='upper center')

# Testing Acccuracy
plot5.plot(TAC_NData, linestyle='dashed', marker='o')
plot5.plot(TAC_CData, linestyle='solid', marker='o')
plot5.set_title('Testing Accuracy')
plot5.legend(['Normal','CLAHE'], loc='lower center')

plt.tight_layout()
plt.savefig('mean_training_all.jpg')
plt.show()
```
