```python
import numpy as np
import glob
import random
import pandas as pd
from keras.preprocessing import image
from keras.models import load_model
import matplotlib.pyplot as plt
```

```python
size_ = 256

model = load_model('C:/Users/HP/TA/BattleGround/MODEL_CL100_64-128-256-fix/clahe/256x256/model_epoch-180.h5')
```

```python
image_path = "C:/Users/HP/TA/BattleGround/dataset-norm/TESTING/PNEUMONIA/BACTERIA-864208-0003.jpeg"
predict_img = image.load_img(image_path, target_size = (size_, size_))
plt.imshow(predict_img)
plt.show()
predict_img = image.img_to_array(predict_img)
predict_img = np.expand_dims(predict_img, axis = 0)

hasil = np.argmax((model.predict(predict_img)), axis=-1)
if(hasil == 0):
    print("Hasil Prediksi : NORMAL")
elif(hasil == 1):
    print("Hasil Prediksi : PNEUMONIA")
else:
    print("Error")
```
