```python
import numpy as np
import pandas as pd
import time
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Activation
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint, Callback, CSVLogger
```

```python
size_ = 64

model = Sequential()

model.add(Conv2D(32, (3, 3), input_shape = (size_, size_, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(64, (3, 3), input_shape = (size_, size_, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(64, (3, 3), input_shape = (size_, size_, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Conv2D(128, (3, 3), input_shape = (size_, size_, 3), activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2, 2)))
model.add(Flatten())

model.add(Dense(units = 2, activation = 'softmax'))

model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
PATH='C:/Users/HP/TA/BattleGround'

getData = ImageDataGenerator()
train_data = getData.flow_from_directory(PATH+'/dataset-clahe/TRAINING', target_size = (size_, size_), batch_size = 32, class_mode = 'categorical')
validation_data = getData.flow_from_directory(PATH+'/dataset-clahe/VALIDATION', target_size = (size_, size_), batch_size = 32, class_mode = 'categorical')
```

```python
class getModelHistoryTime(Callback):
    def on_train_begin(self, logs={}):
        self.times = []
    def on_epoch_begin(self, epoch, logs={}):
        self.epoch_time_start = time.time()
    def on_epoch_end(self, epoch, logs={}):
        self.times.append(time.time() - self.epoch_time_start)


check_point = ModelCheckpoint('model_epoch-{epoch:d}.h5', period=20)
total_time = getModelHistoryTime()
csv_logger = CSVLogger('training_history.csv')
model.fit(x=train_data, epochs=200, callbacks=[check_point,total_time,csv_logger], validation_data=validation_data, steps_per_epoch=1, validation_steps=1)
```

```python
time = np.array(total_time.times)

arr20 = time[:20]
arr40 = time[:40]
arr60 = time[:60]
arr80 = time[:80]
arr100 = time[:100]
arr120 = time[:120]
arr140 = time[:140]
arr160 = time[:160]
arr180 = time[:180]
arr200 = time[:200]

t_arr = [arr20,arr40,arr60,arr80,arr100,arr120,arr140,arr160,arr180,arr200]
pd_arr = pd.DataFrame(t_arr, index=["20-Epoch","40-Epoch","60-Epoch","80-Epoch","100-Epoch","120-Epoch","140-Epoch","160-Epoch","180-Epoch","200-Epoch"])
pd_arr[200] = pd_arr.sum(axis=1)
pd_arr = pd_arr.replace(np.nan, 0)
pd_arr = pd_arr.T
pd_arr.to_csv('training_times.csv', index=False)
```
