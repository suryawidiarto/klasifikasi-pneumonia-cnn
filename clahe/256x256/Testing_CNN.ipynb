```python
import glob
import numpy as np
import pandas as pd
from keras.preprocessing import image
from keras.models import load_model
```

```python
size_ = 256

input_normal = "C:/Users/HP/TA/BattleGround/dataset-clahe/TESTING/NORMAL"
input_pneumonia = "C:/Users/HP/TA/BattleGround/dataset-clahe/TESTING/PNEUMONIA"

model_e20 = load_model('model_epoch-20.h5')
model_e40 = load_model('model_epoch-40.h5')
model_e60 = load_model('model_epoch-60.h5')
model_e80 = load_model('model_epoch-80.h5')
model_e100 = load_model('model_epoch-100.h5')
model_e120 = load_model('model_epoch-120.h5')
model_e140 = load_model('model_epoch-140.h5')
model_e160 = load_model('model_epoch-160.h5')
model_e180 = load_model('model_epoch-180.h5')
model_e200 = load_model('model_epoch-200.h5')
```

```python
#PREDICT NORMAL
e20=[]
e40=[]
e60=[]
e80=[]
e100=[]
e120=[]
e140=[]
e160=[]
e180=[]
e200=[]

for img in glob.glob(input_normal + "/*.jpeg"):
    
    img = image.load_img(img, target_size = (size_, size_))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis = 0)
    
    hasil_e20 = np.argmax((model_e20.predict(img)), axis=-1)
    hasil_e40 = np.argmax((model_e40.predict(img)), axis=-1)
    hasil_e60 = np.argmax((model_e60.predict(img)), axis=-1)
    hasil_e80 = np.argmax((model_e80.predict(img)), axis=-1)
    hasil_e100 = np.argmax((model_e100.predict(img)), axis=-1)
    hasil_e120 = np.argmax((model_e120.predict(img)), axis=-1)
    hasil_e140 = np.argmax((model_e140.predict(img)), axis=-1)
    hasil_e160 = np.argmax((model_e160.predict(img)), axis=-1)
    hasil_e180 = np.argmax((model_e180.predict(img)), axis=-1)
    hasil_e200 = np.argmax((model_e200.predict(img)), axis=-1)
    
    e20.append(hasil_e20)
    e40.append(hasil_e40)
    e60.append(hasil_e60)
    e80.append(hasil_e80)
    e100.append(hasil_e100)
    e120.append(hasil_e120)
    e140.append(hasil_e140)
    e160.append(hasil_e160)
    e180.append(hasil_e180)
    e200.append(hasil_e200)
    
pred_normal = [e20,e40,e60,e80,e100,e120,e140,e160,e180,e200]


#PREDICT PNEUMONIA
e20=[]
e40=[]
e60=[]
e80=[]
e100=[]
e120=[]
e140=[]
e160=[]
e180=[]
e200=[]

for img in glob.glob(input_pneumonia + "/*.jpeg"):
    
    img = image.load_img(img, target_size = (size_, size_))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis = 0)
    
    hasil_e20 = np.argmax((model_e20.predict(img)), axis=-1)
    hasil_e40 = np.argmax((model_e40.predict(img)), axis=-1)
    hasil_e60 = np.argmax((model_e60.predict(img)), axis=-1)
    hasil_e80 = np.argmax((model_e80.predict(img)), axis=-1)
    hasil_e100 = np.argmax((model_e100.predict(img)), axis=-1)
    hasil_e120 = np.argmax((model_e120.predict(img)), axis=-1)
    hasil_e140 = np.argmax((model_e140.predict(img)), axis=-1)
    hasil_e160 = np.argmax((model_e160.predict(img)), axis=-1)
    hasil_e180 = np.argmax((model_e180.predict(img)), axis=-1)
    hasil_e200 = np.argmax((model_e200.predict(img)), axis=-1)
    
    e20.append(hasil_e20)
    e40.append(hasil_e40)
    e60.append(hasil_e60)
    e80.append(hasil_e80)
    e100.append(hasil_e100)
    e120.append(hasil_e120)
    e140.append(hasil_e140)
    e160.append(hasil_e160)
    e180.append(hasil_e180)
    e200.append(hasil_e200)
    
pred_pneumonia = [e20,e40,e60,e80,e100,e120,e140,e160,e180,e200]
```

```python
data_predict = pd.DataFrame(index=["20-Epoch","40-Epoch","60-Epoch","80-Epoch","100-Epoch","120-Epoch","140-Epoch","160-Epoch","180-Epoch","200-Epoch"])

data_predict_normal = pd.DataFrame(pred_normal, index=["20-Epoch","40-Epoch","60-Epoch","80-Epoch","100-Epoch","120-Epoch","140-Epoch","160-Epoch","180-Epoch","200-Epoch"])

data_predict_normal
```

```python
data_predict_pneumonia = pd.DataFrame(pred_pneumonia, index=["20-Epoch","40-Epoch","60-Epoch","80-Epoch","100-Epoch","120-Epoch","140-Epoch","160-Epoch","180-Epoch","200-Epoch"])

data_predict_pneumonia
```

```python
#Diprediksi NORMAL
data_predict['TP'] = (data_predict_normal == 0).sum(axis=1)
data_predict['FP'] = (data_predict_normal == 1).sum(axis=1)

data_predict
```

```python
#Diprediksi PNEUMONIA
data_predict['TN'] = (data_predict_pneumonia == 1).sum(axis=1)
data_predict['FN'] = (data_predict_pneumonia == 0).sum(axis=1)

data_predict
```

```python
#AKURASI
data_predict['Akurasi'] = (data_predict['TP'] + data_predict['TN'])/(data_predict['FP'] + data_predict['FN'] + data_predict['TP'] + data_predict['TN'])

data_predict
```

```python
data_predict.to_csv('training_pred_conf.csv', index=False)
```
